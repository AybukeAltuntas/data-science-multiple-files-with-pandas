## Merge Alıştırması

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

### İç Birleştirme Alıştırması

```python
inner_merged_df = pd.merge(a_df, b_df, on='Country')
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('inner_merge',
    inner_merged_shape=inner_merged_df.shape,
    inner_merged_nulls=sum(inner_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### Sol Birleştirme Alıştırması

```python
left_merged_df = pd.merge(a_df, b_df, on='Country',how="left")
```

```python
from nbresult import ChallengeResult

result = ChallengeResult('left_merge',
    left_merged_shape=left_merged_df.shape,
    left_merged_nulls=sum(left_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### Sağ Birleştirme Alıştırması

```python
right_merged_df = pd.merge(a_df, b_df, on='Country',how="right")
right_merged_df
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('right_merge',
    right_merged_shape=right_merged_df.shape,
    right_merged_nulls=sum(right_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

### Dış Birleştirme Alıştırması

```python
outer_merged_df = pd.merge(a_df, b_df, on='Country',how="outer")
outer_merged_df
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('outer_merge',
    outer_merged_shape=outer_merged_df.shape,
    outer_merged_nulls=sum(outer_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

## Olympic Sports and Medals, 1896-2014

```python
countries_df = pd.read_csv("/Users/aybukealtuntas/data-science-multiple-files-with-pandas/dictionary.csv")
```

```python
summer_df = pd.read_csv("/Users/aybukealtuntas/data-science-multiple-files-with-pandas/summer.csv")
```

```python
summer_df
```

```python
winter_df = pd.read_csv("/Users/aybukealtuntas/data-science-multiple-files-with-pandas/winter.csv")
```

```python
winter_df
```

```python
summer_df = summer_df.rename(columns={"Country": "Code"})
winter_df = winter_df.rename(columns={"Country": "Code"})
```

### Verileri Birleştirme

```python
summer_countries_df = pd.merge(countries_df, summer_df, on="Code", how="inner")
summer_countries_df['Season'] = 'Summer'

winter_countries_df = pd.merge(countries_df, winter_df, on="Code", how="inner")
winter_countries_df['Season'] = 'Winter'

all_df = pd.concat([summer_countries_df, winter_countries_df], ignore_index=True)
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('all_df',
    all_df_shape=all_df.shape,
    all_df_columns=set(all_df.columns)
)
result.write()

print(result.check())
```

### En Başarılı Ülkeler Analizi

```python
filtered_df = all_df[all_df['Year'] >= 1984]
top_10_df = filtered_df.groupby('Country')[['Medal']].count()

top_10_df.columns = ['Medal Count']
top_10_df = top_10_df.sort_values(by='Medal Count', ascending=False).head(10)
print(top_10_df)
```

```python
top_10_df.plot(kind="bar")
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games',
    top_country_1=top_10_df.iloc[0]['Medal Count'],
    top_country_10=top_10_df.iloc[9]['Medal Count']
)
result.write()

print(result.check())
```

```python
filtered_df = all_df[all_df['Year'] >= 1984]

grouped = filtered_df.groupby(['Country', 'Season'])['Medal'].count()
top_10_season_df = grouped.unstack(fill_value=0)
top_10_season_df['Total'] = top_10_season_df['Summer'] + top_10_season_df['Winter']
top_10_season_df = top_10_season_df.sort_values(by='Total', ascending=False).head(10)

print(top_10_season_df)
```

```python
top_10_season_df[['Summer', 'Winter']].plot(kind="bar")
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games_season',
    top_country_season_shape=top_10_season_df.shape,
    top_country_1_summer=top_10_season_df.iloc[0]['Summer'],
    top_country_10_winter=top_10_season_df.iloc[9]['Winter']
)
result.write()

print(result.check())
```

```python
filtered_df = all_df[all_df['Year'] >= 1984]
unique_events_df = filtered_df.drop_duplicates(subset=[
    'Year', 'City', 'Sport', 'Discipline', 'Event', 'Country', 'Medal'
])
top_10_events_df = unique_events_df.groupby('Country')[['Event']].count()

top_10_events_df.columns = ['Event Count']
top_10_events_df = top_10_events_df.sort_values(by='Event Count', ascending=False).head(10)

print(top_10_events_df)
```

```python
top_10_events_df.plot(kind='bar')
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games_event',
    top_country_event_shape=top_10_events_df.shape,
    top_country_1_event=top_10_events_df.iloc[0]['Event Count'],
    top_country_10_event=top_10_events_df.iloc[9]['Event Count']
)
result.write()

print(result.check())
```

```python
top_10_combined = pd.merge(top_10_df, top_10_events_df, left_index=True, right_index=True, how='outer')
top_10_combined = top_10_combined.fillna(0)

top_10_combined = top_10_combined.sort_values(by='Medal Count', ascending=False)

print(top_10_combined)
```

```python
top_10_combined.plot(kind='bar')
```

#### Kodunu kontrol et

```python
from nbresult import ChallengeResult

result = ChallengeResult('olympic_games_combined',
    top_combined_shape=top_10_combined.shape,
    top_combined_1_event=top_10_combined.iloc[0]['Event Count'],
    top_combined_1_medal=top_10_combined.iloc[0]['Medal Count'],
    top_combined_10_event=top_10_combined.iloc[9]['Event Count'],
    top_combined_10_medal=top_10_combined.iloc[9]['Medal Count']
)
result.write()

print(result.check())
```
